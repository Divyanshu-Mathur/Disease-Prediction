```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Training.csv',index_col=False)
```

```python
df.drop(columns='Unnamed: 133',axis=1,inplace=True)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
y = label.fit_transform(df['prognosis'])
```

```python
x=df.drop(columns='prognosis')
```

```python
test = pd.read_csv('Testing.csv')
x_test = test.drop(columns='prognosis')
y_test = label.transform(test['prognosis'])
```

```python
x_train=x
y_train=y
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
```

```python
models = {
    "RandomForestClassifier":RandomForestClassifier(),
    "DecisionTreeClassifier":DecisionTreeClassifier(),
    "XGBClassifier":XGBClassifier(),
    "LogisticRegression":LogisticRegression()
}
```

```python
params = {
    "RandomForestClassifier": {
        "n_estimators": [50, 100, 150],
        "max_depth": [3, 5, 7, 9],
        "min_samples_split": [3, 5, 8]
    },
    "DecisionTreeClassifier": {
        "max_depth": [5, 10, 15],
        "min_samples_split": [2, 4, 6],
        "criterion": ["gini", "entropy"]
    },
    "XGBClassifier": {
        "n_estimators": [100,150, 50],
        "max_depth": [3, 6, 9],
        "learning_rate": [0.01, 1, 0.1]
    },
    "LogisticRegression": {
        "solver": ["lbfgs", "saga"],
        "C": [0.1, 0.5, 1.0, 2.0],
        "penalty": ["l1", "l2"]
    }
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score
report = {}
for i in range(len(models)):
    name = list(models.keys())[i]
    model = list(models.values())[i]
    param = params[name]
    print(f"     Enter {name}       ")
    random = RandomizedSearchCV(estimator=model,param_distributions=param,scoring='accuracy',cv=3,n_iter=50)
    random.fit(x_train,y_train)
    report[name] = random.best_params_
    y_pred = random.predict(x_test)
    print(accuracy_score(y_test, y_pred))
    print(f"     Exit {name}       ")
    
```

```python
report
```

```python
model = DecisionTreeClassifier(min_samples_split=2,max_depth=15,criterion='entropy')
model.fit(x_train,y_train)
```

```python
import joblib
with open('model.pkl','wb') as file:
    joblib.dump(model,file)
```

```python
with open('label.pkl','wb') as file:
    joblib.dump(label,file)
```
